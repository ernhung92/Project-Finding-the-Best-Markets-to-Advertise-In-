## III. New Coders - Locations & Densities

## IV. Spending Money for Learning

We can use the <font color=red>MoneyForLearning</font> column, which describes in American dollars the amount of money spent by participants from the moment they started coding until the moment the completed the survey. 

We can narrow down our analysis to only 4 countries: the US, India, UK, Canada since these countries have the highest absolute frequencies in our sample, and our courses are in English, and English is an official language in all these four countries. 

Let's start with creating a new column that describes the amount of money a student has spent per month so far. To do that, we'll need to divide the <font color=red>MoneyForLearning</font> column to the <font color=red>MonthsProgramming</font> column. The problem is that some students answered that they have been learning to code for 0 months when in reality they have just started

```python
good_respondents["MonthsProgramming"] = good_respondents["MonthsProgramming"].replace(0,1)
good_respondents["MonthlyExpenditure"] = good_respondents["MoneyForLearning"]/good_respondents["MonthsProgramming"]
good_respondents["MonthlyExpenditure"].isnull().value_counts()
```

There are some missing values. Let's remove them

```python
good_respondents = good_respondents[good_respondents["MonthlyExpenditure"].notnull()]
good_respondents["MonthlyExpenditure"].isnull().value_counts()
```

```python
good_respondents["CountryLive"].isnull().value_counts()
```

```python
good_respondents = good_respondents[good_respondents["CountryLive"].notnull()]
good_respondents["CountryLive"].isnull().value_counts()
```

Now we don't have any missing values in our data. Let's find out how much money a student spends on average each month in the US, India, UK, and Canada

```python
good_respondents['CountryLive'].value_counts().head()
```

```python
countries_mean = good_respondents.groupby("CountryLive").mean()
countries_mean['MonthlyExpenditure'].loc[['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

Observations:
1. <font color=red>It's surprising to see that people from India spend more on average per month than people from the UK and Canada</font>

This may be happening due that we don't have enough representative data for the UK, Canada and India, or we have some outliers, making the mean too big for India or too low for the UK and Canada.

## IV. Dealing With Extreme Outliers

```python
import seaborn as sns

good_respondents = good_respondents[(good_respondents["CountryLive"]=="United States of America")|
                                          (good_respondents["CountryLive"]=="India")|
                                               (good_respondents["CountryLive"]=="United Kingdom")|
                                                    (good_respondents["CountryLive"]=="Canada")]
sns.boxplot(x="CountryLive",y="MonthlyExpenditure", data=good_respondents)
```

It's hard to see what is going on on the plots. But we can see that the US has an outlier, or person who spent an average of 50000 USD per month, which is possible, but not a very likely representation of our population. So we need to remove it

```python
good_respondents = good_respondents[good_respondents["MonthlyExpenditure"]<50000]
countries_mean = good_respondents.groupby("CountryLive").mean()
countries_mean['MonthlyExpenditure'].loc[['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

```python
sns.boxplot(x="CountryLive",y="MonthlyExpenditure", data=good_respondents)
```

Now we can see two extreme outliers for India and the US

```python
good_respondents[(good_respondents["MonthlyExpenditure"] > 3000) & (good_respondents["CountryLive"]=="India")]
```

```python
good_respondents[(good_respondents["MonthlyExpenditure"] > 6000) & (good_respondents["CountryLive"]=="United States of America")]
```

```python
countries_mean = good_respondents.groupby("CountryLive").mean()
countries_mean['MonthlyExpenditure'].loc[['United States of America',
                            'India', 'United Kingdom',
                            'Canada']]
```

Obviously, one country we should advertise is the US. Lots of new coders live there and they are willing to pay a good amount of money each month. We want to choose one more market though, but how can we decide between the other three countries?

## V. Finding the Second Best Market Using Z-Scores

Looking at mean table, Canada seems like the rational choice because people there are willing to pay roughly 36 USD per month, compared to India's 30 and UK's 19. 

However, we can use the Z-scores to figure out which of these three markets feel like the least expensive one. To do that, we can do a z-score for 59 USD for each market(since we sell subscriptions at that price), and the market with the lowest z-score will be our choice

```python
second_market_choices = ["India","Canada","United Kingdom"]

for country in second_market_choices:
    which_country = good_respondents[good_respondents["CountryLive"]==country]
    mean = which_country["MonthlyExpenditure"].mean()
    st_dev = which_country["MonthlyExpenditure"].std()
    z_score = (59-mean)/st_dev
    print(country + ": " + str(z_score))
```

Canada has the lowest z-score (0.22), but India's is just slightly greater (0.23). We can stop considering the United Kingdom at this point because its z-score is quite big compared to the other two countries.
With these two z-scores that are very close to one other, it might make sense to actually choose India because there are more potential customers there:
